```python
from scripts.bugsinpy_utils import *
```

```python
python_versions = []

projects = get_projects()
for project in projects:
    bugs = get_bugs(project)
    for bug in bugs:
        info = get_bug_info(project, bug)
        version = info["python_version"]
        version = '.'.join(version.split('.')[:-1])
        if version not in python_versions:
            python_versions.append(version)
```

```python
python_versions
```

```python
for version in python_versions:
    tag_name = f"python-{version}-bullseye"
    print(f"Building Docker image for Python {tag_name}...")
    !docker build --build-arg PYTHON_VERSION={version}-bullseye -f Dockerfile.pythonimages.template -t {tag_name} .
    sif_name = f"images/{tag_name}.sif"
    print(f"Converting Docker image to SIF: {sif_name}...")
    !apptainer build {sif_name} docker-daemon://{tag_name}:latest
```
